```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = load_iris(as_frame=True).frame
pca = PCA(n_components=2)
iris[['new_feature_1', 'new_feature_2']] = pca.fit_transform(iris[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']])
```

```python
iris
```

```python
sns.scatterplot(data=iris, x="sepal length (cm)", y="sepal width (cm)", hue="target", palette=sns.color_palette("bright"))
plt.xlabel("sepal length (cm)", fontweight="bold", fontsize=15)
plt.ylabel("sepal width (cm)", fontweight="bold", fontsize=15)
plt.title("Iris Flower Dataset", fontweight="bold", fontsize=15)
```

```python
sns.scatterplot(data=iris, x="new_feature_1", y="new_feature_2", hue="target", palette=sns.color_palette("bright"))
plt.xlabel("New X1 feature", fontweight="bold", fontsize=15)
plt.ylabel("New X2 feature", fontweight="bold", fontsize=15)
plt.title("PCA dimensionality reduction from 4D to 2D", fontweight="bold", fontsize=15)
```
